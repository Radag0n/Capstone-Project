Our dataset contains several columns with missing values that need to be addressed properly. Let's devise appropriate approaches to handle these missing values in a meaningful manner.

    As a result, we have decided to keep the 'Director2' column in the dataset.

* Notably, 14 movies in the dataset have no certifications.

----  Handling Missing values in Metascore ----

```python
mts = df['Metascore'].isnull()
metas=df[mts]
metas
```

* Metascore is computed by taking the average of the overall movie scores from www.metacritic.com, but it requires a minimum of 4 reviews by critics to generate a score.

* Given that a substantial number of Indian movies fall outside the scope of these critics' reviews, they fail to meet the minimum review criteria, leading to the absence of Metascores for these movies.

* As a result, the missing data for Metascores seems to follow a "Missing Not At Random" (MNAR) pattern, as the absence of scores appears to be related to movies of Indian origin within the dataset.

```python
sns.lmplot(x='Metascore',y='Ratings',data=df)


correlation_coefficient = df['Ratings'].corr(df['Metascore'])
print("Correlation coefficient:", correlation_coefficient)
```

* Upon analyzing the relationship graph and correlation between Ratings and Metascore, it becomes clear that they have a strong positive linear correlation.
* Hence, we can utilize the Ratings values to impute the missing data.

Using Linear Regression Model to impute the missing Metascore values

```python
#Splitting the DataFrame into two parts based on null values.
df_with_metascore = df[df['Metascore'].notnull()]
df_missing_metascore = df[df['Metascore'].isnull()]

#Training a linear regression model using the non-null values
regressor = LinearRegression()
regressor.fit(df_with_metascore[['Ratings']], df_with_metascore['Metascore'])

#Predicting the missing values
predicted_metascore = regressor.predict(df_missing_metascore[['Ratings']])

#Assignin min max Metascores because it cannot be negative or zero
min_metascore = 7
max_metascore = 90
predicted_metascore = predicted_metascore.clip(min_metascore, max_metascore)

# Fill the missing 'Metascore' values with the predicted values
df_missing_metascore['Metascore'] = predicted_metascore

# Concatenate the two parts of datafarme back togethre
imputed_df = pd.concat([df_with_metascore, df_missing_metascore])


#Converting them to whole numbers
imputed_df['Metascore'] = imputed_df['Metascore'].round().astype(int)

imputed_df
```

```python
#renaming imputed df into movies_df
movies_df=imputed_df
```

* Our objective is to predict "Gross Collection" and the number of votes for movies.

* "Gross Collection" (in millions) is the target variable (dependent variable).

* Imputing missing values in the target variable is not advisable because it can influence the learning algorithm's behavior.

* As a result, we will exclude movies with missing "Gross Collection" values from our prediction model for gross collection.

* However, we will include these movies in the prediction model for the number of votes.

* WIth this approach, we preserve the robustness of our gross collection prediction model while still using movies with missing gross collection data for predicting the number of votes.

---- Handling Certification missing values ----

```python
movies_df['Certification'].value_counts()
```

The "Certification" column contains various certification values that often overlap due to different certification names used in different countries. To simplify this, we will group and categorize movie ratings as G, PG, PG-13, R, Rejected.

```python
# Mapping the certification ratings of movies
rating_mapping = {
    'U': 'G',
    'PG': 'PG',
    '9+': 'PG',
    'UA': 'PG',
    '12': 'PG-13',
    '12A': 'PG-13',
    'PG-13': 'PG-13',
    'AA': 'PG-13',
    '15': 'R',
    '15+': 'R',
    '16': 'R',
    '18': 'R',
    'A': 'R',
    'R': 'R',
    'X': 'X',
    'Rejected': 'Rejected',
    'Not Rated': 'Not Rated',
}

movies_df['Certification'] = movies_df['Certification'].map(rating_mapping)
```

```python
cert = movies_df['Certification'].isnull()
cert_missing=df[cert]
cert_missing
```

* Out of the total 1775 movies in the dataset, 14 movies (0.78%) have no recorded Certification information.

* The Certification column contains categorical data, representing different movie certifications.

* Instead of imputing the missing Certification values, we will introduce a new category in the Certification column labeled as 'Not Available' to represent the missing values.

```python
movies_df['Certification']=movies_df['Certification'].fillna('Not Available')
```

```python
movies_df['Certification'].value_counts()
```

---- Handling Star Columns missing values

* The 'Star2' and 'Star3' columns each have one missing value, while the 'Star4' column has two missing values.
* The 'Stars' columns contain categorical data, representing the names of actors in the movies.
* Rather than imputing the missing star names, we will introduce a new category in the 'Stars' columns labeled as 'Not Available' to represent the missing values.

```python
star_missing_columns=['Star2','Star3','Star4']
for i in star_missing_columns:
    movies_df[i]=movies_df[i].fillna('Not Available')
```

---- Handling Genre Columns missing values

* The Genre2 column have one missing value and Genre 3 volumn has 148 missing values.
* Similarly like other catergorical data columns, we will replace the missing values as 'Not Available'.

```python
genre_missing_columns=['Genre2','Genre3']
for i in genre_missing_columns:
    movies_df[i]=movies_df[i].fillna('Not Available')
```
